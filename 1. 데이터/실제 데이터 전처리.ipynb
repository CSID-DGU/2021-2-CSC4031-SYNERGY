```python
X1 = data[data['투입 후 운전용량'].notnull()]['모선전압'].values.reshape(-1,1)
y1 = data[data['투입 후 운전용량'].notnull()]['투입 후 운전용량'].values
```

```python
X2 = data[data['투입 후 운전용량'].isnull()]['모선전압'].values.reshape(-1,1)
y2 = data[data['투입 후 운전용량'].isnull()]['투입 전 운전용량'].values
```

```python
X = np.vstack([X1, X2])
X.shape
```

```python
y = np.hstack([y1, y2])
y.shape
```

- 전압과 투입용량만 고려  
- 투입 후 운전용량이 NaN이면 투입 전 운전용량을 고려

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor

xgb = XGBRegressor()

xgb.fit(X, y)
```

```python
pred = xgb.predict(X)
```

```python
mean_squared_error(y, pred)
```
